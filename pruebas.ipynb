```python
import matplotlib.pyplot as plt
import random

x = [random.uniform(0, 100) for _ in range(50)]
y = [random.uniform(0, 100) for _ in range(50)]

plt.scatter(x, y, marker='o', color='blue', label='Random Scatter Plot')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

plt.legend()

plt.show()
```
